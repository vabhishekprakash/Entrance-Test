```python
import pandas as pd
import sqlite3
```

```python
orders = pd.read_csv("orders.csv")
orders.head()
```

```python
users = pd.read_json("users.json")
users.head()
```

```python

conn = sqlite3.connect("restaurants.db")
cursor = conn.cursor()

with open("restaurants.sql", encoding="utf-8") as f:
    sql_data = f.read()

cursor.executescript(sql_data)
restaurants = pd.read_sql_query("SELECT * FROM restaurants", conn)
restaurants.head()
```

```python
orders_users = pd.merge(
    orders,
    users,
    on="user_id",
    how="left"
)

orders_users.head()
```

```python
final_data = pd.merge(
    orders_users,
    restaurants,
    on="restaurant_id",
    how="left"
)

final_data.head()
```

```python
final_data.to_csv("final_food_delivery_dataset.csv", index=False)

print("Final file created!")
```

```python
df = final_data.copy()

df['order_date'] = pd.to_datetime(df['order_date'], format="%d-%m-%Y")
```

```python
#Q4
def rating_range(r):
    if 3.0 <= r <= 3.5: return "3.0-3.5"
    if 3.6 <= r <= 4.0: return "3.6-4.0"
    if 4.1 <= r <= 4.5: return "4.1-4.5"
    if 4.6 <= r <= 5.0: return "4.6-5.0"

df['rating_range'] = df['rating'].apply(rating_range)
df.groupby('rating_range')['total_amount'] \
.sum() \
.sort_values(ascending=False)

```
